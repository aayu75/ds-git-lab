```python
import pandas as pd
df = pd.read_csv("/content/titanic.csv")
```

```python
#how many passengers out of total passengers have "paid_more_than_median" set to 1
```

```python
# introducing new column
median_fare = df['Fare'].median()
df['paid_more_than_median'] = (df['Fare'] > median_fare).astype(int)

print(df.head())
```

```python
# Count how many passengers paid more than median
count = df['paid_more_than_median'].sum()
total = len(df)

print("Passengers who paid more than median fare:", count)
print("Total passengers:", total)
```
